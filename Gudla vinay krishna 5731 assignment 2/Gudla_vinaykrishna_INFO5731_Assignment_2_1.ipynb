# Question 2 (30 points)

Write a python program to **clean the text data** you collected in the previous question and save the clean data in a new column in the csv file. The data cleaning steps include: [Code and output is required for each part]

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the stopwords list.

(4) Lowercase all texts

(5) Stemming.

(6) Lemmatization.

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Load the dataset
df = pd.read_csv("filtered_amazon_reviews.csv")

# Define a function to clean the text
def clean_text(text):
    # Remove special characters and punctuations
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Tokenize text
    tokens = word_tokenize(text)

    # Remove stopwords
    tokens = [word for word in tokens if word.lower() not in stopwords.words('english')]

    # Lowercase all texts
    tokens = [word.lower() for word in tokens]

    # Stemming
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(word) for word in tokens]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    # Join tokens back to string
    cleaned_text = ' '.join(tokens)
    return cleaned_text

# Apply the cleaning function to the review_body column
df['cleaned_review_body'] = df['review_body'].apply(clean_text)

# Save the dataframe with the new column to a new CSV file
df.to_csv('cleaned_amazon_reviews.csv', index=False)

print("Data cleaning complete and saved to cleaned_amazon_reviews.csv")
```

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
import nltk

# Sample text for demonstration
sample_text = "This is a sample review that i am testing ! It contains numbers 123 and special characters @#&*$. Also, stopwords."

# # Ensure NLTK resources are downloaded
# nltk.download('stopwords')
# nltk.download('punkt')
# nltk.download('wordnet')

print("Original Text:")
print(sample_text)

# 1. Remove noise, such as special characters and punctuations.
text_no_noise = re.sub(r'[^a-zA-Z\s]', '', sample_text)
print("\nAfter removing noise:")
print(text_no_noise)

# 2. Remove numbers.
text_no_numbers = re.sub(r'\d+', '', text_no_noise)
print("\nAfter removing numbers:")
print(text_no_numbers)

# Tokenize text for further processing
tokens = word_tokenize(text_no_numbers)

# 3. Remove stopwords.
tokens_no_stopwords = [word for word in tokens if word.lower() not in stopwords.words('english')]
print("\nAfter removing stopwords:")
print(' '.join(tokens_no_stopwords))

# 4. Lowercase all texts.
tokens_lowercase = [word.lower() for word in tokens_no_stopwords]
print("\nAfter converting to lowercase:")
print(' '.join(tokens_lowercase))

# 5. Stemming.
stemmer = PorterStemmer()
tokens_stemmed = [stemmer.stem(word) for word in tokens_lowercase]
print("\nAfter stemming:")
print(' '.join(tokens_stemmed))

# 6. Lemmatization.
lemmatizer = WordNetLemmatizer()
tokens_lemmatized = [lemmatizer.lemmatize(word) for word in tokens_stemmed]
print("\nAfter lemmatization:")
print(' '.join(tokens_lemmatized))
```

```python
# Your code here
import spacy
import nltk
from spacy import displacy
import benepar

# Load spaCy's English language model
nlp = spacy.load("en_core_web_sm")

# Enable benepar for constituency parsing
benepar.download('benepar_en3')
nlp.add_pipe('benepar', config={'model': 'benepar_en3'})

# Sample clean text for demonstration
clean_text = "Apple Inc. announced the iPhone 12 in California on October 13, 2020. It features innovative technology."

# Processing text with spaCy for NER and dependency parsing
doc = nlp(clean_text)

# (1) POS Tagging and counting
pos_counts = {'NOUN': 0, 'VERB': 0, 'ADJ': 0, 'ADV': 0}
for token in doc:
    if token.pos_ in pos_counts:
        pos_counts[token.pos_] += 1
print("POS Counts:", pos_counts)

# (2) Dependency Parsing (Using spaCy's visualization)
displacy.render(doc, style='dep', jupyter=True, options={'distance': 90})

# Constituency Parsing (Using benepar)
sentence = list(doc.sents)[0]  # Taking the first sentence for demonstration
constituency_parse = sentence._.parse_string
print("Constituency Parsing Tree:", constituency_parse)

# (3) Named Entity Recognition and counting entities
entity_counts = {}
for ent in doc.ents:
    if ent.label_ in entity_counts:
        entity_counts[ent.label_] += 1
    else:
        entity_counts[ent.label_] = 1
print("Named Entities and Counts:", entity_counts)

```
